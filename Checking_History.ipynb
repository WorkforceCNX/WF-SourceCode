```python
# Function for scanning folder and get the data
def ImportProperties(DF,folder):
        for file in glob.glob(folder):
            #get file name
            filename = os.path.basename(file) 
            #get folder name
            Folder_name = os.path.basename(os.path.dirname(folder))
            #get modified time
            time_mod = os.path.getmtime(file)
            local_time_mod = dt.fromtimestamp(time_mod).strftime('%m/%d/%Y, %I:%M:%S %p')
            #find out who modified if possible
            try:
                compressfile = zipfile.ZipFile(file)
                readXML = xml.dom.minidom.parseString(compressfile.read('docProps/core.xml')).toprettyxml(indent='  ')

                asText = readXML.splitlines()
                for item in asText:
                    if 'lastModifiedBy' in item:
                        itemLength = len(item)-20
                        Modified_by = item[21:itemLength]
            except:
                Modified_by = 'N/A'
            #Data collected
            temp_list =[Folder_name,filename,local_time_mod,Modified_by]
            DF=pd.concat([DF,pd.DataFrame([temp_list],columns=['Folder Name','File','Last Modified Date','Last Modified By'])],join='outer')  
        return DF
```

```python
#MAIN - Just run all 

#Scan data
NewExcel = pd.DataFrame(columns=['Folder Name','File','Last Modified Date',\
                                     '2nd to Last Modified Date',\
                                        '3rd to Last Modified Date',\
                                            '4th to Last Modified Date',\
                                                '5th to Last Modified Date','Last Modified By'])
for folder in folder_paths:
        NewExcel = ImportProperties(NewExcel,folder)

#1st run setup
if not os.path.exists(Output_file):
    NewExcel.to_excel(Output_file, index=False)
#Onward Run
else:
    Current_Excel = pd.read_excel(Output_file,header=0)
    Shift_DF = Current_Excel.iloc[:,[2,3,4,5,6]].shift(periods=1,axis='columns')

    Current_Excel[['Last Modified Date',\
                                     '2nd to Last Modified Date',\
                                        '3rd to Last Modified Date',\
                                            '4th to Last Modified Date',\
                                                '5th to Last Modified Date']] = Shift_DF
    NewExcel['Identify'] = NewExcel['Folder Name'] +''+NewExcel['File']
    Current_Excel['Identify'] = Current_Excel['Folder Name'] +''+Current_Excel['File']
    
    #Update
    Current_Excel = Current_Excel.merge(NewExcel[['Identify', 'Last Modified Date','Last Modified By']],on='Identify',how="left",suffixes=('', '_right'))
    Current_Excel[['Last Modified Date','Last Modified By']] = Current_Excel[['Last Modified Date_right','Last Modified By_right']]

    Current_Excel.drop(['Last Modified Date_right','Last Modified By_right'],inplace=True,axis=1)
    #Check deleted
    Current_Excel.loc[:,'Last Modified Date']=Current_Excel.loc[:,'Last Modified Date'].fillna("Deleted")
    #New Entry
    Current_Excel = pd.concat([Current_Excel,NewExcel])
    Current_Excel.drop_duplicates(subset='Identify',inplace=True)
    Current_Excel.drop(['Identify'],inplace=True,axis=1)
    
    #Extract
    Current_Excel.to_excel(Output_file, index=False)
```
