```python
#FUNCTION TO IMPORT FOLDERS

dtype_error_map = [['float64', 'int64'], ['datetime64[ns]', 'object'], ['object', 'float64'], ['int64', 'object']]
go = 'y'


def import_csv(path):
    list_dir = sorted([x for x in os.listdir(path) if x.endswith('.csv')])
    raw = []
    try:
        standard_cols = pd.read_csv(os.path.join(path, list_dir[0])).columns.to_list()
        standard_dtype = pd.read_csv(os.path.join(path, list_dir[0])).dtypes.to_dict()
    except:
        standard_cols = pd.read_csv(os.path.join(path, list_dir[0]), encoding='latin-1').columns.to_list()
        standard_dtype = pd.read_csv(os.path.join(path, list_dir[0]), encoding='latin-1').dtypes.to_dict()
    change_log = []
    global go
    for file in list_dir:
        file_path = os.path.join(path, file)
        try:
            file_data = pd.read_csv(file_path)
        except:
            file_data = pd.read_csv(file_path, encoding='latin-1')
        file_data_dtype =  file_data.dtypes.to_dict()
        for col in file_data.columns.to_list(): #check unexpected columns
            if col not in standard_cols:
                print(f'{file} contains unexpected column {col}')
                change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "unexpected column", col])
        for col in standard_cols: #check missing columns
            if col not in file_data.columns.to_list():
                print(f'{file} has a missing column {col}')
                change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "missing column", col])
        for key in standard_dtype:
            if key in file_data.columns.to_list() and standard_dtype[key] != file_data_dtype[key]:
                if [standard_dtype[key], file_data_dtype[key]] in dtype_error_map or [file_data_dtype[key], standard_dtype[key]] in dtype_error_map:
                    pass
                else:
                    print(f'{file} {key}:{standard_dtype[key]} -> {file_data_dtype[key]}')
                    change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "wrong dtype", key, f"{standard_dtype[key]}->{file_data_dtype[key]}"])
        file_data["filename"] = file
        raw.append(file_data)

    final_raw = pd.concat(raw, ignore_index=True)
    if len(change_log)>0:
        go = input(f'Do you want to proceed with the changes/defects? (y/n)').lower()
        if go == 'y':
              with open(raw_log, 'a', newline='') as file:
                   for change in change_log:
                        csv.writer(file).writerow(change) 
        else:
            sys.exit()           
    return final_raw
    
def import_xlsx(path, **kwargs):
    list_dir = sorted([x for x in os.listdir(path) if x.endswith('.xlsx')])
    raw = []
    standard_cols = pd.read_excel(os.path.join(path, list_dir[0]), engine="openpyxl", sheet_name = list(kwargs.values())[0]).columns.to_list()
    standard_dtype = pd.read_excel(os.path.join(path, list_dir[0]), engine="openpyxl", sheet_name = list(kwargs.values())[0]).dtypes.to_dict()
    change_log = []
    global go
    for file in list_dir:
        file_path = os.path.join(path, file)
        try:
            next = True
            sheetname_position = 0
            while next == True and sheetname_position<= len(kwargs.values()):
                try: 
                    file_data = pd.read_excel(file_path, engine="openpyxl", sheet_name = list(kwargs.values())[sheetname_position])
                    next = False
                except:

                    sheetname_position += 1
                
            file_data_dtype =  file_data.dtypes.to_dict()
            for col in file_data.columns.to_list(): #check unexpected columns
                if col not in standard_cols:
                    print(f'{file} contains unexpected column {col}')
                    change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "unexpected column", col])
            for col in standard_cols: #check missing columns
                if col not in file_data.columns.to_list():
                    print(f'{file} has a missing column {col}')
                    change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "missing column", col])
            for key in standard_dtype:
                if key in file_data.columns.to_list() and standard_dtype[key] != file_data_dtype[key]:
                    if [standard_dtype[key], file_data_dtype[key]] in dtype_error_map or [file_data_dtype[key], standard_dtype[key]] in dtype_error_map:
                        pass
                    else:
                        print(f'{file} {key}:{standard_dtype[key]} -> {file_data_dtype[key]}')
                        change_log.append([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, file, "wrong dtype", key])
            file_data["filename"] = file
            raw.append(file_data)
        except:
            print(f'Cannot input {file} into dataframe')

    final_raw = pd.concat(raw, ignore_index=True)
    if len(change_log)>0:
        go = input(f'Do you want to proceed with the changes/defects? (y/n)').lower()
        if go == 'y':
            with open(raw_log, 'a', newline='') as file:
                for change in change_log:
                    csv.writer(file).writerow(change)
    return final_raw
```

```python
def check_histo(file_json_path, new_file):
    asis_file = pd.read_json(file_json_path)
    tobe_file = new_file
    loop_col_dict ={
        'masterroster.json': ['Employee_ID', 'GEO', 'Site_ID', 'Employee_Last_Name', 'Employee_First_Name', 'Status', 'Wave #', 'Role', 'Booking Login ID', 'Language Start Date', 'TED Name', 'CUIC Name', 'PST_Start_Date', 'Production_Start_Date', 'LWD', 'Termination_Date', 'Designation', 'cnx_email', 'Booking Email', 'WAH Category', 'Full name', 'IEX', 'filename'],
        'eps.json': ['EID', 'BPE Code'],
        'eps_full.json': ['EID', 'Session login date','BPE Code'],
        'aht.json': ['Date', 'Staff', 'Language', 'Item Channel', 'Topic', 'Subtopic'],
        'aht_full.json': ['Date', 'Staff', 'Language', 'Item Channel', 'Topic', 'Subtopic'],
        'csat.json': ['Date', 'Team', 'Channel', 'Staff', 'Type', 'Topic of the first Ticket', 'Language', 'Csat 2.0 Score'],
        'csat_full.json': ['Date', 'Team', 'Channel', 'Staff', 'Type', 'Topic of the first Ticket', 'Language', 'Csat 2.0 Score'],
        'csat_reso.json': ['Date', 'Team', 'Channel', 'Staff', 'Type', 'Topic of the first Ticket', 'Language', 'Csat 2.0 Score'],
        'csat_reso_full.json': ['Date', 'Team', 'Channel', 'Staff', 'Type', 'Topic of the first Ticket', 'Language', 'Csat 2.0 Score'],
        'ramco.json': ['EID', 'Value', 'Employee_type'],
        'ramco_full.json': ['EID', 'Value', 'Employee_type'],
        'schedule.json': ['Emp ID', 'Name', 'Date', 'Shift', 'LOB', 'Shift_type', 'team_leader', 'week_shift', 'week_off', 'OM', 'DPE','Check PO'],
        'schedule_full.json': ['Emp ID', 'Name', 'Date', 'Shift', 'LOB', 'Shift_type', 'team_leader', 'week_shift', 'week_off', 'OM', 'DPE','Check PO'],
        'cuic.json': ['Employee_ID', 'Shift_type', 'Session Date'],
        'cuic_full.json': ['Employee_ID', 'Shift_type', 'Session Date'],
        'exceptional_hrs.json': ['Emp ID', 'Date', 'TL', 'OM'],
        'iex.json': ['ID Agent', 'EID'],
        'ioshrinkage.json': ['LOB', 'Week', 'Ratio', 'filename'],
        'ot.json': ['Emp ID', 'Date', 'OT'],
        'ot_full.json': ['Emp ID', 'Date', 'OT'],
        'psat.json': ['Date', 'Staff Name', 'Language', 'Final Topics', 'How satisfied were you with our service?', 'How difficult did we make it or you to solve your issue?', 'Agent understood my question', 'Agent did everything possible to help me', 'Did we fully resolve your issue?', 'Channel', 'filename'],
        'psat_full.json': ['Date', 'Staff Name', 'Language', 'Final Topics', 'How satisfied were you with our service?', 'How difficult did we make it or you to solve your issue?', 'Agent understood my question', 'Agent did everything possible to help me', 'Did we fully resolve your issue?', 'Channel', 'filename'],
        'quality.json': ['eval_date', 'agent_username', 'evaluator_username', 'result', 'final_question_grouping', 'template_group', 'eval_template_name', 'sections', 'sitecode', 'eval_language', 'tix_final_topic', 'tix_final_subtopic', 'csat_language_code', 'csat_satisfied', 'filename'],
        'quality_full.json': ['eval_date', 'agent_username', 'evaluator_username', 'result', 'final_question_grouping', 'template_group', 'eval_template_name', 'sections', 'sitecode', 'eval_language', 'tix_final_topic', 'tix_final_subtopic', 'csat_language_code', 'csat_satisfied', 'filename'],
        'ramcoot.json': ['employee_code', 'employee_name', 'Employee Type', 'OT Type', 'Date', 'Status', 'Hours', 'filename'],
        'ramcoot_full.json': ['employee_code', 'employee_name', 'Employee Type', 'OT Type', 'Date', 'Status', 'Hours', 'filename'],
        'requirement.json': ['LOB', 'Date', 'Daily Requirement', 'Prod Requirement', 'filename'],
        'workplan_merge_iex.json': ['LOB', 'Date', 'Agent ID', 'Agent Name', 'Scheduled Activity', 'EID'],
        'labels.json': ['Attendance Code', 'Description', 'Type', 'Schedule', 'Actual', 'Planned leaves', 'Actual leaves', 'filename'],
        'kpi_target.json': ['LOB', 'LOB Group', 'Week', 'Tenure days'],
        'cpi.json': ['Staff Name', 'Employee_ID', 'Date', 'Channel'],
        'cpi_full.json': ['Staff Name', 'Employee_ID', 'Date', 'Channel'],
        'rona.json': ['Employee_ID', 'Session Date'],
        'rona_full.json': ['Employee_ID', 'Session Date'],
        'workplan_wst.json': ['LOB', 'Agent ID', 'Date', 'date_starttime', 'date_endtime', 'Schedule Act','date_duration','Length', 'EID'],
        'csat_pega.json':['Survey Id', 'Reservation', 'Team', 'Channel','Staff', 'Type', 'Date','Topic of the first Ticket','Language','Csat 2.0 Score','Employee_ID'],
        'csat_pega_full.json':['Survey Id', 'Reservation', 'Team', 'Channel','Staff', 'Type', 'Date','Topic of the first Ticket','Language','Csat 2.0 Score','Employee_ID'],
        'iph.json':["Staff Name","Employee_ID","Date","Channel"],
        'iph_full.json':["Staff Name","Employee_ID","Date","Channel"]}
        
    def dict_converter(file):
        if os.path.basename(file_json_path) in loop_col_dict.keys():
            cols = [x for x in loop_col_dict[os.path.basename(file_json_path)]]
        else:
            cols = file.columns.to_list()

        the_dict = {}
        # loop through all columns in this file
        for col in cols:
            values = file[col].drop_duplicates().to_list()
            # remove Null values
            values = [x for x in values if not (isinstance(x, float) and np.isnan(x)) and x is not None]
                          
            the_dict[col] = values
        return the_dict

    # dictionaries for historical data and newly updated data below:
    histo_dict = dict_converter(file=asis_file)
    new_dict = dict_converter(file=tobe_file)

    # a list of common keys of the two dicts
    histo_dict_keys = [x for x in histo_dict.keys() if x in new_dict.keys()]
    new_dict_keys = [x for x in new_dict.keys() if x in histo_dict.keys()]

    # check missing values
    for key in histo_dict_keys:
        for value in histo_dict[key]:
            try:
                if value not in new_dict[key]:
                    print(f"missing value found {key}: {value}")
                    with open(raw_log, 'a', newline='') as file:
                        csv.writer(file).writerow([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, os.path.basename(file_json_path), "missing values", key, value])
            except:
                print(f"error-missing {key}: {value}")
    # check for new values
    for key in new_dict_keys:
        for value in new_dict[key]:
            try:
                if value not in histo_dict[key]:
                    print(f"newly updated value found {key}: {value}")
                    with open(raw_log, 'a', newline='') as file:
                        csv.writer(file).writerow([dt.now().strftime("%Y-%m-%d %H:%M:%S"), personal_path, os.path.basename(file_json_path), "newly updated value", key, value])     
            except:
                print(f"error-newly updated {key}: {value}")
    return histo_dict
```

```python
# IMPORT WORKPLAN WITH START TIME
if go == 'y':
    workplan_wst = import_xlsx(path=workplan_wst_path, sheet_name='Sheet1')
    # remove rows where there's no activity
    workplan_wst = workplan_wst.loc[workplan_wst['Start2'].notna()]

    workplan_wst['Date_end'] = workplan_wst.apply(lambda x: pd.to_datetime(x['Date'], format='mixed') + pd.Timedelta(days=1) if x['Start1'] >= t(15, 0, 0) else pd.to_datetime(x['Date'], format='mixed'), axis=1)
    workplan_wst['date_starttime'] = pd.to_datetime(workplan_wst['Date'], format='mixed').dt.strftime('%Y-%m-%d') + " " + workplan_wst['Start1'].astype('str')
    workplan_wst['date_starttime'] = pd.to_datetime(workplan_wst['date_starttime'], format='mixed')
    workplan_wst['date_endtime'] = pd.to_datetime(workplan_wst['Date_end'], format='mixed').dt.strftime('%Y-%m-%d') + " " + workplan_wst['End1'].astype('str')
    workplan_wst['date_endtime'] = pd.to_datetime(workplan_wst['date_endtime'], format='mixed')

    workplan_wst['date_act_start'] = workplan_wst.apply(lambda x: pd.to_datetime(x['Date'], format='mixed') + pd.Timedelta(days=1) if x['Start1'] >= t(20, 0, 0) and x['Start2'] <= t(15, 0, 0) else pd.to_datetime(x['Date'], format='mixed'),axis=1)
    workplan_wst['act_starttime'] = pd.to_datetime(workplan_wst['date_act_start'], format='mixed').dt.strftime('%Y-%m-%d') + " " + workplan_wst['Start2'].astype('str')
    workplan_wst['act_starttime'] = pd.to_datetime(workplan_wst['act_starttime'], format='mixed')
    workplan_wst['date_act_end'] = workplan_wst.apply(lambda x: pd.to_datetime(x['Date'], format='mixed') + pd.Timedelta(days=1) if x['Start1'] >= t(20, 0, 0) and x['End2'] <= t(15, 0, 0) else pd.to_datetime(x['Date'], format='mixed'),axis=1)
    workplan_wst['act_endtime'] = pd.to_datetime(workplan_wst['date_act_end'], format='mixed').dt.strftime('%Y-%m-%d') + " " + workplan_wst['End2'].astype('str')
    workplan_wst['act_endtime'] = pd.to_datetime(workplan_wst['act_endtime'], format='mixed')

    # date_duration show the total time that the agents are workplanned to work in a day
    workplan_wst['date_duration'] = (pd.to_datetime(workplan_wst['date_endtime'], format='%Y-%m-%d%H:%M:%S', errors='coerce') - pd.to_datetime(workplan_wst['date_starttime'], format='%Y-%m-%d%H:%M:%S', errors='coerce'))
    workplan_wst['date_duration'] = workplan_wst.apply(lambda x: x['date_duration'].total_seconds()/86400, axis=1)
    workplan_wst['date_duration'] = workplan_wst['date_duration'].round(2)

    # act_duration show the total time that the agents are workplanned to work in a day
    workplan_wst['act_duration'] = (pd.to_datetime(workplan_wst['act_endtime'], format='%Y-%m-%d%H:%M:%S', errors='coerce') - pd.to_datetime(workplan_wst['act_starttime'], format='%Y-%m-%d%H:%M:%S', errors='coerce'))
    workplan_wst['act_duration'] = workplan_wst.apply(lambda x: x['act_duration'].total_seconds()/86400, axis=1)
    workplan_wst['act_duration'] = workplan_wst['act_duration'].round(2)

    # Trim the table
    workplan_wst = workplan_wst[['LOB', 'ID', 'Date', 'date_starttime', 'date_endtime', 'Schedule Act','act_starttime','act_endtime', 'date_duration', 'act_duration']]
    # Rename ID to Agent ID
    workplan_wst = workplan_wst.rename(columns={'ID': 'Agent ID', 'act_duration': 'Length'})

    # Get ID (CNX ID)
    workplan_wst = pd.merge(workplan_wst, masterroster[['IEX','Employee_ID']], left_on='Agent ID', right_on ='IEX', how='left')
    workplan_wst = workplan_wst.rename(columns={'Employee_ID': 'EID'})
    workplan_wst = workplan_wst.drop(columns=['IEX','LOB'])
    
    #Get LOB from Roster
    schedule['Date']=pd.to_datetime(schedule['Date'],format='mixed').dt.date
    workplan_wst['Date']=pd.to_datetime(workplan_wst['Date'],format='mixed').dt.date    
    workplan_wst = pd.merge(workplan_wst,schedule[['Emp ID','Date','LOB']], left_on=['EID','Date'], right_on =['Emp ID','Date'], how='left')    
    workplan_wst['Date'] = pd.to_datetime(workplan_wst['Date'], errors='coerce')
    workplan_wst = workplan_wst.drop(columns=['Emp ID'])

    # convert datetime value to string and export to json
    workplan_wst['date_starttime']=pd.to_datetime(workplan_wst['date_starttime'], format='mixed').dt.strftime('%Y-%m-%d %I:%M:%S %p')
    workplan_wst['date_endtime']=pd.to_datetime(workplan_wst['date_endtime'], format='mixed').dt.strftime('%Y-%m-%d %I:%M:%S %p')
    workplan_wst['act_starttime'] = pd.to_datetime(workplan_wst['act_starttime'], format='mixed').dt.strftime('%Y-%m-%d %I:%M:%S %p')
    workplan_wst['act_endtime'] = pd.to_datetime(workplan_wst['act_endtime'], format='mixed').dt.strftime('%Y-%m-%d %I:%M:%S %p')

    check_histo(file_json_path=workplan_wst_output_path, new_file=workplan_wst)
    if go == 'y':
        workplan_wst.to_json(workplan_wst_output_path, orient='records', date_format='iso')
```
